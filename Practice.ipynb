```python
# You have a DataFrame with employee information, and you need to calculate a bonus based on specific criteria:

# Employees in the "Finance" or "IT" departments.
# Employees with more than 5 years of experience.
# Employees with a performance rating of "A" or "B".
# The bonus is calculated as follows:

# If an employee meets all the above criteria, their bonus is 15% of their shub-salary.
# If they do not meet all the criteria but still have a performance rating of "A", they get a bonus of 5% 
# of their shub-salary.

# # Calculate Bonus:

# For employees meeting all three criteria, the bonus should be 15% of shub-salary.
# For employees who only have a performance rating of "A", the bonus should be 5% of shub-salary.
# **Create a new column Bonus in the DataFrame to store the calculated bonuses.
```

```python
# Initialize Bonus column with 0
df['Bonus'] = 0

# Calculate Bonus for employees meeting all criteria (15% of shub-salary)
df.loc[
    (df['Department'].isin(['Finance', 'IT'])) & 
    (df['Experience'] > 5) & 
    (df['Performance'].isin(['A', 'B'])),
    'Bonus'
] = df['Salary'] * 0.15

# Calculate Bonus for employees with a performance rating of "A" (5% of shub-salary)
df.loc[
    (df['Performance'] == 'A') & 
    ~((df['Department'].isin(['Finance', 'IT'])) & 
    (df['Experience'] > 5)),
    'Bonus'
] = df['Salary'] * 0.05
```
