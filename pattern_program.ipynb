```python
#not mine
n = 5  # Number of rows

for i in range(1, n + 1):  # Loop to control rows
    # Print leading spaces
    for j in range(n, i, -1):  # Decreasing spaces
        print(" ", end="")
    
    # Print increasing numbers
    for j in range(1, i + 1):  # Increasing part of numbers
        print(j, end="")
    
    # Print decreasing numbers
    for j in range(i - 1, 0, -1):  # Decreasing part of numbers
        print(j, end="")
    
    print()  # Move to the next line
```
